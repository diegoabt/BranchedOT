```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
import sys

sys.path.append('../../ready functions in py/')

%load_ext autoreload
%autoreload 2
```

```python
sizes = np.array([10,20,30,50,70,100,150])
mean_growth_cost_arr = np.zeros(len(sizes))
mean_growthMC_cost_arr = np.zeros(len(sizes))
mean_stress_cost_arr = np.zeros(len(sizes))
mean_stressMC_cost_arr = np.zeros(len(sizes))

std_growth_cost_arr = np.zeros(len(sizes))
std_growthMC_cost_arr = np.zeros(len(sizes))
std_stress_cost_arr = np.zeros(len(sizes))
std_stressMC_cost_arr = np.zeros(len(sizes))

mean_MC_iter_arr = np.zeros(len(sizes))
mean_growth_iter_arr = np.zeros(len(sizes))
mean_stress_iter_arr = np.zeros(len(sizes))

std_MC_iter_arr = np.zeros(len(sizes))
std_growth_iter_arr = np.zeros(len(sizes))
std_stress_iter_arr = np.zeros(len(sizes))


for i,size in enumerate(sizes):
    pkl_file_path = f"results/heuristic_comparison_size{size}_full.pkl"
    pkl_file = open(pkl_file_path, 'rb')
    results_dict = pickle.load(pkl_file)
    pkl_file.close()
    
    all_growth_cost_arr = np.array([])
    all_growthMC_cost_arr = np.array([])
    all_stress_cost_arr = np.array([])
    all_stressMC_cost_arr = np.array([])
    
    all_MC_iter_arr = np.array([])
    all_growth_iter_arr = np.array([])
    all_stress_iter_arr = np.array([])
    
    for thread_dict in results_dict:
        for key in thread_dict:
            all_growth_cost_arr = np.append(all_growth_cost_arr, thread_dict[key]["growth_cost"]/thread_dict[key]["MC_cost"])
            all_growthMC_cost_arr = np.append(all_growthMC_cost_arr, thread_dict[key]["growth_cost_MC"]/thread_dict[key]["MC_cost"])
            all_stress_cost_arr = np.append(all_stress_cost_arr, thread_dict[key]["stress_cost"]/thread_dict[key]["MC_cost"])
            all_stressMC_cost_arr = np.append(all_stressMC_cost_arr, thread_dict[key]["stress_cost_MC"]/thread_dict[key]["MC_cost"])
            
            all_MC_iter_arr = np.append(all_MC_iter_arr, thread_dict[key]["MC_iterations"])
            all_growth_iter_arr = np.append(all_growth_iter_arr, thread_dict[key]["growth_MC_iterations"])
            all_stress_iter_arr = np.append(all_stress_iter_arr, thread_dict[key]["stress_MC_iterations"])
    
    mean_growth_cost_arr[i] = np.mean(all_growth_cost_arr)
    std_growth_cost_arr[i] = np.std(all_growth_cost_arr)
    mean_growthMC_cost_arr[i] = np.mean(all_growthMC_cost_arr)
    std_growthMC_cost_arr[i] = np.std(all_growthMC_cost_arr)
    
    mean_stress_cost_arr[i] = np.mean(all_stress_cost_arr)
    std_stress_cost_arr[i] = np.std(all_stress_cost_arr)
    mean_stressMC_cost_arr[i] = np.mean(all_stressMC_cost_arr)
    std_stressMC_cost_arr[i] = np.std(all_stressMC_cost_arr)
    
    mean_MC_iter_arr[i] = np.mean(all_MC_iter_arr)
    std_MC_iter_arr[i] = np.std(all_MC_iter_arr)  
    mean_growth_iter_arr[i] = np.mean(all_growth_iter_arr)
    std_growth_iter_arr[i] = np.std(all_growth_iter_arr)
    mean_stress_iter_arr[i] = np.mean(all_stress_iter_arr)
    std_stress_iter_arr[i] = np.std(all_stress_iter_arr)
```

```python
# just to get the stuff for the stress right:
mean_stress_cost_arr = np.zeros(len(sizes))
mean_stressMC_cost_arr = np.zeros(len(sizes))

std_stress_cost_arr = np.zeros(len(sizes))
std_stressMC_cost_arr = np.zeros(len(sizes))

mean_stress_iter_arr = np.zeros(len(sizes))
std_stress_iter_arr = np.zeros(len(sizes))


for i,size in enumerate(sizes):
    pkl_file_path = f"results/heuristic_comparison_size{size}_redone.pkl"
    pkl_file = open(pkl_file_path, 'rb')
    results_dict = pickle.load(pkl_file)
    pkl_file.close()
    
    all_stress_cost_arr = np.array([])
    all_stressMC_cost_arr = np.array([])
    all_stress_iter_arr = np.array([])
    
    for thread_dict in results_dict:
        for key in thread_dict:
            all_stress_cost_arr = np.append(all_stress_cost_arr, thread_dict[key]["stress_cost"]/thread_dict[key]["MC_cost"])
            all_stressMC_cost_arr = np.append(all_stressMC_cost_arr, thread_dict[key]["stress_cost_MC"]/thread_dict[key]["MC_cost"])
            
            all_stress_iter_arr = np.append(all_stress_iter_arr, thread_dict[key]["stress_MC_iterations"])
    
    mean_stress_cost_arr[i] = np.mean(all_stress_cost_arr)
    std_stress_cost_arr[i] = np.std(all_stress_cost_arr)
    mean_stressMC_cost_arr[i] = np.mean(all_stressMC_cost_arr)
    std_stressMC_cost_arr[i] = np.std(all_stressMC_cost_arr)
    
    mean_stress_iter_arr[i] = np.mean(all_stress_iter_arr)
    std_stress_iter_arr[i] = np.std(all_stress_iter_arr)
```

```python
plt.errorbar(sizes, mean_growth_cost_arr, yerr=std_growth_cost_arr, fmt=".-", capsize=3, label="ICH")
plt.errorbar(sizes, mean_stress_cost_arr, yerr=std_stress_cost_arr, fmt=".-", capsize=3, label="IMST")
plt.errorbar(sizes, mean_growthMC_cost_arr, yerr=std_growthMC_cost_arr, fmt=".-", capsize=3, label="ICH-SA")
plt.errorbar(sizes, mean_stressMC_cost_arr, yerr=std_stressMC_cost_arr, fmt=".-", capsize=3, label="IMST-SA")
plt.grid()
plt.legend()
plt.xticks(sizes, sizes)
plt.ylabel("average cost ratio")
plt.xlabel("number of terminals $n$")
plt.savefig("cost-final.pdf")
plt.show()
```

```python
plt.errorbar(sizes, mean_growth_iter_arr, yerr=std_growth_iter_arr, fmt=".-", capsize=3, label="ICH")
plt.errorbar(sizes, mean_stress_iter_arr, yerr=std_stress_iter_arr, fmt=".-", capsize=3, label="IMST")
plt.errorbar(sizes, mean_MC_iter_arr, yerr=std_MC_iter_arr, fmt=".-", capsize=3, label="star topology")
plt.grid()
plt.legend()
plt.xticks(sizes, sizes)
plt.ylabel("average iterations until convergence")
plt.xlabel("number of terminals $n$")
#plt.savefig("iterations-final.pdf")
plt.show()
```

```python
# comparison with respect to alpha:
segments = 20

all_growth_cost_arr = np.zeros((segments, len(sizes) * 150))
all_growthMC_cost_arr = np.zeros((segments, len(sizes) * 150))
all_stress_cost_arr = np.zeros((segments, len(sizes) * 150))
all_stressMC_cost_arr = np.zeros((segments, len(sizes) * 150))

count = 0
for size in sizes:
    pkl_file_path = f"results/heuristic_comparison_size{size}_full.pkl"
    pkl_file = open(pkl_file_path, 'rb')
    results_dict = pickle.load(pkl_file)
    pkl_file.close()
    
    for thread_dict in results_dict:
        for key in thread_dict:
            seg = int(segments * thread_dict[key]["bot_problem_dict"]["al"])
            
            all_growth_cost_arr[seg, count] =  thread_dict[key]["growth_cost"]/thread_dict[key]["MC_cost"]
            all_growthMC_cost_arr[seg, count] = thread_dict[key]["growth_cost_MC"]/thread_dict[key]["MC_cost"]
            all_stress_cost_arr[seg, count] = thread_dict[key]["stress_cost"]/thread_dict[key]["MC_cost"]
            all_stressMC_cost_arr[seg,count] = thread_dict[key]["stress_cost_MC"]/thread_dict[key]["MC_cost"]
            count += 1

mean_growth_cost_seg_arr = np.zeros(segments)
mean_growthMC_cost_seg_arr = np.zeros(segments)
mean_stress_cost_seg_arr = np.zeros(segments)
mean_stressMC_cost_seg_arr = np.zeros(segments)

std_growth_cost_seg_arr = np.zeros(segments)
std_growthMC_cost_seg_arr = np.zeros(segments)
std_stress_cost_seg_arr = np.zeros(segments)
std_stressMC_cost_seg_arr = np.zeros(segments)

for seg in range(segments):
    mask = (all_growth_cost_arr[seg,:] > 0)
    print(sum(mask))

    mean_growth_cost_seg_arr[seg] = np.mean(all_growth_cost_arr[seg,mask])
    std_growth_cost_seg_arr[seg] = np.std(all_growth_cost_arr[seg,mask])
    mean_growthMC_cost_seg_arr[seg] = np.mean(all_growthMC_cost_arr[seg,mask])
    std_growthMC_cost_seg_arr[seg] = np.std(all_growthMC_cost_arr[seg,mask])
    
    mean_stress_cost_seg_arr[seg] = np.mean(all_stress_cost_arr[seg,mask])
    std_stress_cost_seg_arr[seg] = np.std(all_stress_cost_arr[seg,mask])
    mean_stressMC_cost_seg_arr[seg] = np.mean(all_stressMC_cost_arr[seg,mask])
    std_stressMC_cost_seg_arr[seg] = np.std(all_stressMC_cost_arr[seg,mask])
```
