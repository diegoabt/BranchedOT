```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cost(a,x,y,weights):
    cost = np.zeros_like(x)
    for i in range(len(a)):
        cost += weights[i] * np.sqrt((a[i,0]-x)**2 + (a[i,1] -y)**2)
    return cost
```

```python
a = np.array([
    [0.5,0.1],
    [0.2,0.9],
    [0.8,0.9]
])

al = 0.7
flows = np.array([1,0.5,0.5]) 

x = np.linspace(0,1,1000)
y = x

x,y = np.meshgrid(x,y)
costi = cost(a,x,y,weights=flows**al)
i = np.argmin(costi)
x_min = x.reshape(-1)[i]
y_min = y.reshape(-1)[i]

plt.figure(figsize=(8,6.5))
plt.contourf(x,y, costi, cmap="Blues", levels=30, zorder=-1)
cbar = plt.colorbar()
cbar.ax.tick_params(labelsize=13)

for i in range(len(a)):
    plt.plot([x_min, a[i,0]],[y_min, a[i,1]], "-", linewidth=2, color="black", zorder=0.5)

plt.scatter(a[0,0],a[0,1], marker ="+", linewidth=4, s = 500, color="red", zorder=1)
#plt.plot(a[0,0],a[0,1], marker = "+", color = "r", linestyle="", markersize=15, markeredgewidth = 5,  label="sources")
plt.scatter(a[1,0],a[1,1], marker="_", linewidth=4, s = 500, color="blue", zorder=1)
plt.scatter(a[2,0],a[2,1], marker="_", linewidth=4, s = 500, color="blue", zorder=1)
plt.xlim(0,1)
plt.ylim(0,1)
plt.xticks(np.linspace(0,1,3), fontsize=13)
plt.yticks(np.linspace(0,1,3), fontsize=13)
plt.axis("equal")
plt.savefig("contour-y.pdf", bbox_inches="tight")
plt.show()
```
