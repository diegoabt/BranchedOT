```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
m_arr = np.arange(10) + 1
num_terminals_arr = np.array([10,20,30,50,70,100,150])
for num_terminals in num_terminals_arr:
    pkl_file_path = f"results/incremental_m_check_probs300_size{num_terminals}.pkl"
    pkl_file = open(pkl_file_path, 'rb')
    data_dict = pickle.load(pkl_file)
    pkl_file.close()
    
    cost_m_arr = np.zeros(len(m_arr))
    for thread_dict in data_dict:
        for key in thread_dict:
            if thread_dict[key]["bot_problem_dict"]["al"] < 2: #no alpha constraitn
                cost_m_arr += thread_dict[key]["cost_m_arr"]
    
    
    #plot the rescaled cost_m_arr:
    plt.title(f"Summed cost of incremental growth ({num_terminals} terminals)")
    plt.plot(m_arr, cost_m_arr, ".-", label=f"{num_terminals}", alpha=0.7)
    plt.legend()
    plt.show()
```
