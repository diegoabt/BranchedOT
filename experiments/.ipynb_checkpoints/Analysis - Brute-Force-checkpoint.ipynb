```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
import pickle
import sys
import time
from tqdm import tqdm

%load_ext autoreload
%autoreload 2

sys.path.append('../helper functions/')

from helper_fcts import *
from iterative_BOT_solver import iterative_bot_solver
from mc_update import monte_carlo_step
from interpolating_prior import interpolated_prior
```

```python
# open pickle:
N = 100

# store them in a pickle file:
pkl_file_path = f"brute-force_results{N}.pkl"
pkl_file = open(pkl_file_path, 'rb')
data_dict = pickle.load(pkl_file)
pkl_file.close()
```

```python
plot = True

for key in data_dict:

    solution = data_dict[key]
    bot_problem_dict = solution["bot_problem_dict"]
    overall_best_topo = solution["overall_best_topo"]
    overall_best_cost = solution["overall_best_cost"]
    cost_arr = solution["overall_cost_arr"]
    
    al = bot_problem_dict["al"]
    coords_sources = bot_problem_dict["coords_sources"]
    coords_sinks = bot_problem_dict["coords_sinks"]
    supply_arr = bot_problem_dict["supply_arr"]
    demand_arr = bot_problem_dict["demand_arr"]
    
    if len(supply_arr) + len(demand_arr) != 9:
        continue
    
    print("\n\nal=",al)
    
    best_cost, coords_iter = iterative_bot_solver(overall_best_topo, supply_arr, demand_arr, coords_sources, coords_sinks, al,
                                         relative_improvement_threshold=1e-6, min_iterations=-1,
                                         max_iterations=2000,
                                         plot=plot, title=f"$\\alpha =${np.round(al,2)}", fov=None, save=True, save_name=f"n9/img{2 * key}")
    
    
    plt.hist(cost_arr, bins=30)
    plt.axvline(np.mean(cost_arr), color="orange", label="mean")
    plt.axvline(np.median(cost_arr), color="darkred", label="median")
    plt.ylabel("Counts")
    plt.xlabel("Cost for all topologies")
    plt.legend(fontsize=13)
    plt.savefig(f"n9/img{2*key + 1}a.pdf")
    plt.show()
    
    continue 
    
    # compare to MC heuristic:
    # init star graph:
    topo = nx.Graph()
    for node in range(len(supply_arr) + len(demand_arr)):
        topo.add_edge(-1, node)

    cost, coords_iter = iterative_bot_solver(topo, supply_arr, demand_arr, coords_sources, coords_sinks, al,
                                             relative_improvement_threshold=1e-6, min_iterations=-1,
                                             max_iterations=1000,
                                             plot=False, title="", fov=None, save=False, save_name="img")

    # MC iterations:
    T = 0
    keep_going = True
    iteration = 0
    sample_edge_list = list(topo.edges())
    while keep_going and iteration < 60000:
        iteration += 1
        topo, sample_edge_list, cost, coords_iter, accepted = monte_carlo_step(topo, sample_edge_list, cost,
                                        coords_iter, bot_problem_dict, temperature=T)

        # check if converged:
        if len(sample_edge_list) == 0:
            # this means that we have stuck with one topology for an entire run through.
            keep_going = False

    # plt.plot(np.arange(len(cost_arr)), cost_arr)
    # plt.show()
    _ = iterative_bot_solver(topo, supply_arr, demand_arr, coords_sources, coords_sinks, al,
                         relative_improvement_threshold=1e-6, min_iterations=-1,
                         max_iterations=2000,
                         plot=plot, title="", fov=None, save=False, save_name="img")
    
    plt.hist(cost_arr, bins=30)
    plt.axvline(np.mean(cost_arr), color="orange", label="mean")
    plt.axvline(np.median(cost_arr), color="darkred", label="median")
    plt.ylabel("Counts")
    plt.xlabel("Cost for all topologies")
    plt.legend()
    plt.show()
    
    print("best_cost=", overall_best_cost, np.min(cost_arr), "cost=", cost)
```

```python
#evaluate the performance for different sizes:
np.random.seed(178)

plot = False
N = 100

ratio_arr_star = np.zeros(6*N)
ratio_arr_random = np.zeros(6*N)
iteration_arr = np.array([])

for key in tqdm(data_dict):
    if key == N*6:
        break
    solution = data_dict[key]
    bot_problem_dict = solution["bot_problem_dict"]
    overall_best_topo = solution["overall_best_topo"]
    overall_best_cost = solution["overall_best_cost"]
    cost_arr = solution["overall_cost_arr"]
    
    al = bot_problem_dict["al"]
    coords_sources = bot_problem_dict["coords_sources"]
    coords_sinks = bot_problem_dict["coords_sinks"]
    supply_arr = bot_problem_dict["supply_arr"]
    demand_arr = bot_problem_dict["demand_arr"]
    num_terminals = len(supply_arr) + len(demand_arr)
    
    # compare to MC heuristic, star:
    # init star graph:
    #topo = nx.Graph()
    #for node in range(len(supply_arr) + len(demand_arr)):
    #    topo.add_edge(-1, node)

    #cost, coords_iter = iterative_bot_solver(topo, supply_arr, demand_arr, coords_sources, coords_sinks, al,
    #                                         relative_improvement_threshold=1e-6, min_iterations=-1,
    #                                         max_iterations=1000,
    #                                         plot=False, title="", fov=None, save=False, save_name="img")
    
    # USe mST as init tree toplogy:
    topo = interpolated_prior(bot_problem_dict, beta = 1)
    cost, coords_iter = iterative_bot_solver(topo, supply_arr, demand_arr, coords_sources, coords_sinks, al,
                                             relative_improvement_threshold=1e-6, min_iterations=-1,
                                             max_iterations=1000,
                                             plot=False, title="", fov=None, save=False, save_name="img")

    T = 0
    keep_going = True
    iteration = 0
    sample_edge_list = list(topo.edges())
    while keep_going and iteration < 60000:
        iteration += 1
        topo, sample_edge_list, cost, coords_iter, accepted = monte_carlo_step(topo, sample_edge_list, cost,
                                        coords_iter, bot_problem_dict, temperature=T)

        # check if converged:
        if len(sample_edge_list) == 0:
            # this means that we have stuck with one topology for an entire run through.
            keep_going = False

    
    cost_star,_ = iterative_bot_solver(topo, supply_arr, demand_arr, coords_sources, coords_sinks, al,
                         relative_improvement_threshold=1e-6, min_iterations=-1,
                         max_iterations=2000,
                         plot=plot, title="", fov=None, save=False, save_name="img")
    
    if num_terminals == 9:
        iteration_arr = np.append(iteration_arr, iteration)
    
    #cost ratios:
    ratio_arr_star[key] =  cost_star / overall_best_cost
    continue
    
    # compare to MC heuristic, random:
    # init star graph:
    topo = generate_random_binary_tree_topo(num_terminals)

    cost, coords_iter = iterative_bot_solver(topo, supply_arr, demand_arr, coords_sources, coords_sinks, al,
                                             relative_improvement_threshold=1e-6, min_iterations=-1,
                                             max_iterations=1000,
                                             plot=False, title="", fov=None, save=False, save_name="img")
    
    T = 0
    keep_going = True
    iteration = 0
    sample_edge_list = list(topo.edges())
    while keep_going and iteration < 60000:
        iteration += 1
        topo, sample_edge_list, cost, coords_iter, accepted = monte_carlo_step(topo, sample_edge_list, cost,
                                        coords_iter, bot_problem_dict, temperature=T)

        # check if converged:
        if len(sample_edge_list) == 0:
            # this means that we have stuck with one topology for an entire run through.
            keep_going = False

    _ = iterative_bot_solver(topo, supply_arr, demand_arr, coords_sources, coords_sinks, al,
                         relative_improvement_threshold=1e-6, min_iterations=-1,
                         max_iterations=2000,
                         plot=plot, title="", fov=None, save=False, save_name="img")
    
    cost_random,_ = iterative_bot_solver(topo, supply_arr, demand_arr, coords_sources, coords_sinks, al,
                         relative_improvement_threshold=1e-7, min_iterations=-1,
                         max_iterations=2000,
                         plot=plot, title="", fov=None, save=False, save_name="img")
    
    if num_terminals == 9:
        iteration_arr = np.append(iteration_arr, iteration)
    
    #cost ratios:
    ratio_arr_star[key] =  cost_star / overall_best_cost
    ratio_arr_random[key] =  cost_random / overall_best_cost
    
ratio_arr_star = ratio_arr_star.reshape((6,N))
ratio_arr_random = ratio_arr_random.reshape((6,N))
```

```python
ratio_arr = np.clip(ratio_arr_star, 1.0, None)
```
